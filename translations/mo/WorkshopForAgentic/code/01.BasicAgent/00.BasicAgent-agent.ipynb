```python
from agent_framework.ollama import OllamaChatClient
```

```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
agent = OllamaChatClient(model_id=os.getenv("OLLAMA_CHAT_MODEL_ID")).create_agent(
        name="PodcastAgent",
        instructions="""
        您是我的播客中文脚本生成助手。请根据提供的内容，生成10分钟中文播客脚本。
        播客脚本需要注意由主持人Lucy和专家Ken共同主持。脚本内容根据输入的内容产生，最后输出格式如下：

            speaker 1：…… \n
            speaker 2：…… \n
            speaker 1：…… \n
            speaker 2：…… \n
            speaker 1：…… \n
            speaker 2：…… \n
            """)
```

```python
prompt = """

人工智能（Artificial Intelligence，简称AI）是计算机科学的一个分支，旨在开发能够模拟人类智能行为的机器或软件。其核心目标是使机器具备感知、学习、推理、决策和创造等能力，以完成需要人类智慧的任务。以下是人工智能的关键特点和应用领域：

---

### **1. 核心能力**
- **学习能力**：通过数据训练模型，自动改进性能（如推荐系统）。
- **推理与逻辑**：解决复杂问题（如棋类游戏的策略分析）。
- **感知能力**：通过视觉（图像识别）、听觉（语音识别）等感知环境。
- **语言理解**：自然语言处理（NLP），支持机器翻译、聊天机器人等。
- **自主决策**：在动态环境中做出最优选择（如自动驾驶）。

---

### **2. 类型分类**
- **弱人工智能（Narrow AI）**：专注于特定任务，如 Siri、人脸识别、医疗影像分析。
- **强人工智能（General AI）**：具备与人类相当的通用认知能力（目前尚未实现）。
- **超级人工智能（Super AI）**：超越人类智能，仍处于理论阶段。

---

### **3. 技术基础**
- **机器学习（ML）**：通过算法从数据中学习规律（如深度学习、强化学习）。
- **深度学习（DL）**：利用神经网络模拟人脑结构，擅长处理图像、语音等非结构化数据。
- **自然语言处理（NLP）**：使机器理解并生成人类语言（如聊天机器人、文本摘要）。

---

### **4. 应用场景**
- **医疗**：疾病诊断、药物研发。
- **金融**：风险评估、量化交易。
- **交通**：自动驾驶、交通流量优化。
- **教育**：个性化学习推荐。
- **娱乐**：游戏AI、内容生成。

---

### **5. 挑战与伦理**
- **数据隐私**：AI依赖大量数据，可能引发隐私泄露。
- **算法偏见**：训练数据的偏差可能导致歧视性决策。
- **就业影响**：自动化可能取代部分传统职业。
- **安全风险**：如深度伪造（Deepfake）技术的滥用。

---

### **6. 发展历程**
- **1956年**：达特茅斯会议首次提出“人工智能”概念。
- **1980年代**：专家系统兴起，但因局限性衰退。
- **2000年后**：大数据和算力提升推动AI爆发，深度学习成为主流。

---

如果需要更具体的解释或案例，可以进一步提问！ 😊

"""
```

```python
response = await agent.run(prompt)
```
