```python
import asyncio

from agent_framework import TextReasoningContent
from agent_framework.ollama import OllamaChatClient
```

```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
agent = OllamaChatClient(model_id=os.getenv("OLLAMA_CHAT_MODEL_ID")).create_agent(
        name="ReasoningAgent",
        instructions="You are my math teacher. Help the student solve the problems step by step.",
        additional_chat_options={"think": True},  # Enable Reasoning on agent level
)
```

```python
query = "解释一下什么是勾股定理"
```

```python
result = await agent.run(query)
```

```python
reasoning = "".join((c.text or "") for c in result.messages[-1].contents if isinstance(c, TextReasoningContent))
```
