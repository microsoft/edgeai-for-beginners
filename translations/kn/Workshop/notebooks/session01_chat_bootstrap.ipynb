# ಸೆಷನ್ 1 – ಚಾಟ್ ಬೂಟ್‌ಸ್ಟ್ರಾಪ್ (ಫೌಂಡ್ರಿ ಲೋಕಲ್)

ಈ ನೋಟ್ಬುಕ್ ಫೌಂಡ್ರಿ ಲೋಕಲ್ ಅನ್ನು ಬೂಟ್‌ಸ್ಟ್ರಾಪ್ ಮಾಡುತ್ತದೆ, ಇಚ್ಛಿತ ಮಾದರಿ ಅಲಿಯಾಸ್ ಅನ್ನು ಡೌನ್‌ಲೋಡ್ ಮಾಡುತ್ತದೆ, ಮತ್ತು ಸಾಮಾನ್ಯ ಮತ್ತು ಸ್ಟ್ರೀಮಿಂಗ್ ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆಯನ್ನು ಎರಡನ್ನೂ ನಿರ್ವಹಿಸುತ್ತದೆ.


# ದೃಶ್ಯ
ಈ ಅಧಿವೇಶನವು Foundry Local ಮೂಲಕ ಸ್ಥಳೀಯ ಸಣ್ಣ ಭಾಷಾ ಮಾದರಿಯನ್ನು ಪ್ರತಿಕ್ರಿಯಿಸಲು ಅಗತ್ಯವಿರುವ ಕನಿಷ್ಠವನ್ನು ಪರಿಚಯಿಸುತ್ತದೆ. ನೀವು:
- SDK / ಕ್ಲೈಂಟ್ ಅವಲಂಬನೆಗಳನ್ನು ಸ್ಥಾಪಿಸುವಿರಿ.
- ಆಯ್ಕೆಮಾಡಿದ ಅಲಿಯಾಸ್ (ಡೀಫಾಲ್ಟ್: `phi-4-mini`)ಗಾಗಿ Foundry Local ನಿರ್ವಾಹಕವನ್ನು ಪ್ರಾರಂಭಿಸುವಿರಿ.
- ಮಾದರಿ ಮೆಟಾಡೇಟಾದಲ್ಲಿ ಐಚ್ಛಿಕ ಕ್ಷೇತ್ರಗಳನ್ನು ಸಹಿಸುವುದಕ್ಕಾಗಿ ರಕ್ಷಣಾತ್ಮಕ ಮಾಂಕಿಪ್ಯಾಚ್ ಅನ್ನು ಅನ್ವಯಿಸುವಿರಿ.
- ಸಾಮಾನ್ಯ ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆ ವಿನಂತಿಯನ್ನು ಕಳುಹಿಸುವಿರಿ.
- ಪ್ರತಿಕ್ರಿಯೆ ಟೋಕನ್-ಬೈ-ಟೋಕನ್ ಸ್ಟ್ರೀಮ್ ಮಾಡುವಿರಿ.

ಗುರಿ ನಿಮ್ಮ ಸ್ಥಳೀಯ ರನ್‌ಟೈಮ್ ಮತ್ತು ನೆಟ್‌ವರ್ಕ್ ಮಾರ್ಗವನ್ನು ಪರಿಶೀಲಿಸುವುದು, ನಂತರ RAG, ರೌಟಿಂಗ್ ಅಥವಾ ಏಜೆಂಟ್‌ಗಳಿಗೆ ಸಾಗುವುದು.


### ವಿವರಣೆ: ಅವಲಂಬನೆ ಸ್ಥಾಪನೆ
ಈ ಕನಿಷ್ಠ ಚಾಟ್ ಫ್ಲೋಗೆ ಅಗತ್ಯವಿರುವ Python ಪ್ಯಾಕೇಜ್‌ಗಳನ್ನು ಸ್ಥಾಪಿಸುತ್ತದೆ:
- `foundry-local-sdk`: ಸ್ಥಳೀಯ ಮಾದರಿಗಳು ಮತ್ತು ಸೇವಾ ಜೀವನಚಕ್ರವನ್ನು ನಿರ್ವಹಿಸಿ.
- `openai`: ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆಗಳಿಗೆ ಪರಿಚಿತ ಕ್ಲೈಂಟ್ ಅವಲೋಕನ.
- `rich`: ಸ್ಪಷ್ಟವಾದ ನೋಟ್ಬುಕ್ ಔಟ್‌ಪುಟ್‌ಗಾಗಿ ಸುಂದರ ಮುದ್ರಣ.

ಮತ್ತೆ ಓಡಿಸುವುದು ಸುರಕ್ಷಿತವಾಗಿದೆ (ಐಡಂಪೊಟೆಂಟ್). ನಿಮ್ಮ ಪರಿಸರದಲ್ಲಿ ಇವುಗಳಿದ್ದರೆ ಬಿಟ್ಟುಬಿಡಿ.

### ವಿವರಣೆ: ಕೋರ್ ಆಮದುಗಳು
ನೋಟ್‌ಬುಕ್‌ನಲ್ಲಿ ಬಳಸಲಾಗುವ ಮೋಡ್ಯೂಲ್‌ಗಳನ್ನು ತರಿಸುತ್ತದೆ:
- ಸ್ಥಳೀಯ ಮಾದರಿ ರನ್‌ಟೈಮ್‌ನೊಂದಿಗೆ ಇಂಟರ್ಫೇಸ್ ಮಾಡಲು `FoundryLocalManager`.
- ಪರಿಚಿತ ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವ API ಸರ್ಫೇಸ್ ಅನ್ನು ಮರುಬಳಕೆ ಮಾಡಲು `OpenAI` ಕ್ಲೈಂಟ್.
- ಶೈಲಿಯುತ ಔಟ್‌ಪುಟ್‌ಗಾಗಿ `rich.print`.

ಇಲ್ಲಿ ಯಾವುದೇ ನೆಟ್‌ವರ್ಕ್ ಕರೆಗಳು ನಡೆಯುವುದಿಲ್ಲ—ಇದು ಕೇವಲ ನಾಮಸ್ಥಳವನ್ನು ಸಿದ್ಧಪಡಿಸುತ್ತದೆ.


```python
import os
from foundry_local import FoundryLocalManager
from foundry_local.models import FoundryModelInfo
from openai import OpenAI
from rich import print
```

### ವಿವರಣೆ: ಮ್ಯಾನೇಜರ್ ಪ್ರಾರಂಭ ಮತ್ತು ಮೆಟಾಡೇಟಾ ಪ್ಯಾಚ್
ಆಯ್ದ ಅಲಿಯಾಸ್‌ಗಾಗಿ `FoundryLocalManager` ಅನ್ನು ಪ್ರಾರಂಭಿಸಿ ಮತ್ತು `promptTemplate` ಶೂನ್ಯವಾಗಿರಬಹುದಾದ ಸೇವಾ ಪ್ರತಿಕ್ರಿಯೆಗಳನ್ನು ಸೌಮ್ಯವಾಗಿ ನಿರ್ವಹಿಸಲು ರಕ್ಷಣಾತ್ಮಕ ಮಾಂಕಿಪ್ಯಾಚ್ ಅನ್ನು ಅನ್ವಯಿಸುತ್ತದೆ.

ಪ್ರಮುಖ ಫಲಿತಾಂಶಗಳು:
- ಸೇವಾ ಸ್ಥಿತಿ ಮತ್ತು ಎಂಡ್‌ಪಾಯಿಂಟ್ ಅನ್ನು ದೃಢೀಕರಿಸುತ್ತದೆ.
- ಕ್ಯಾಶ್ ಮಾಡಲಾದ ಮಾದರಿಗಳನ್ನು ಪಟ್ಟಿ ಮಾಡುತ್ತದೆ (ಸ್ಥಳೀಯ ಸಂಗ್ರಹಣೆಯನ್ನು ಪರಿಶೀಲಿಸುತ್ತದೆ).
- ಅಲಿಯಾಸ್‌ಗೆ ಕಂಕ್ರೀಟ್ ಮಾದರಿ ID ಅನ್ನು ಪರಿಹರಿಸುತ್ತದೆ (ನಂತರದ ಚಾಟ್ ಕರೆಗಳಲ್ಲಿ ಬಳಸಲಾಗುತ್ತದೆ).

ನೀವು ಕಚ್ಚಾ ಸೇವಾ ಮೆಟಾಡೇಟಾದಲ್ಲಿ ಮಾನ್ಯತೆ ಸಮಸ್ಯೆಗಳನ್ನು ಎದುರಿಸಿದರೆ, ಈ ಮಾದರಿ SDK ಅನ್ನು ಫೋರ್ಕ್ ಮಾಡದೆ ಹೇಗೆ ಶುದ್ಧೀಕರಿಸಬಹುದು ಎಂಬುದನ್ನು ತೋರಿಸುತ್ತದೆ.


```python
# Monkeypatch to tolerate service responses where promptTemplate is null
_original_from_list_response = FoundryModelInfo.from_list_response

def _safe_from_list_response(response):  # type: ignore
    try:
        if isinstance(response, dict) and response.get("promptTemplate") is None:
            # Normalize to empty dict so pydantic validation passes
            response["promptTemplate"] = {}
    except Exception as e:  # pragma: no cover
        print(f"[yellow]Warning: safe wrapper encountered issue normalizing promptTemplate: {e}[/yellow]")
    return _original_from_list_response(response)

# Apply patch only once
if getattr(FoundryModelInfo.from_list_response, "__name__", "") != "_safe_from_list_response":
    FoundryModelInfo.from_list_response = staticmethod(_safe_from_list_response)  # type: ignore

ALIAS = os.getenv('FOUNDRY_LOCAL_ALIAS', 'phi-4-mini')
manager = FoundryLocalManager(ALIAS)
print(f'[bold green]Service running:[/bold green] {manager.is_service_running()}')
print(f'Endpoint: {manager.endpoint}')
print('Cached models:', manager.list_cached_models())
model_id = manager.get_model_info(ALIAS).id
print(f'Using model id: {model_id}')
```

### ವಿವರಣೆ: ಮೂಲ ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆ
ಸ್ಥಳೀಯ Foundry ಎಂಡ್‌ಪಾಯಿಂಟ್‌ಗೆ ಸೂಚಿಸುವ `OpenAI`-ಸಮ್ಮತ ಕ್ಲೈಂಟ್ ಅನ್ನು ರಚಿಸಿ ಮತ್ತು ಒಂದು ಏಕ-ನಾನ್‌ಸ್ಟ್ರೀಮಿಂಗ್ ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆಯನ್ನು ನಿರ್ವಹಿಸುತ್ತದೆ. ಇಲ್ಲಿ ಗಮನಿಸಿ:
- ಮಾದರಿ ದೋಷವಿಲ್ಲದೆ ಪ್ರತಿಕ್ರಿಯಿಸುವುದನ್ನು ಖಚಿತಪಡಿಸಿಕೊಳ್ಳಿ.
- ವಿಳಂಬ / ಔಟ್‌ಪುಟ್ ಸ್ವರೂಪವನ್ನು ಪರಿಶೀಲಿಸಿ.
- ಸಂಪನ್ಮೂಲಗಳನ್ನು ಉಳಿಸಲು `max_tokens` ಅನ್ನು ಸಡಿಲವಾಗಿ ಇಡಿ.

ಇದು ವಿಫಲವಾದರೆ, Foundry ಸ್ಥಳೀಯ ಸೇವೆ ಚಾಲನೆಯಲ್ಲಿದೆಯೇ ಮತ್ತು ಅಲಿಯಾಸ್ ಸರಿಯಾಗಿ ಪರಿಹರಿಸಲಾಗಿದೆಯೇ ಎಂದು ಮರುಪರಿಶೀಲಿಸಿ.


```python
client = OpenAI(base_url=manager.endpoint, api_key=manager.api_key or 'not-needed')
prompt = 'List two benefits of local inference for privacy.'
resp = client.chat.completions.create(
    model=model_id,
    messages=[{'role':'user','content':prompt}],
    max_tokens=120,
    temperature=0.5
)
print(resp.choices[0].message.content)
```

### ವಿವರಣೆ: ಸ್ಟ್ರೀಮಿಂಗ್ ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆ
ಮೇಲ್ವಿಚಾರಿತ ತಡತೆಯನ್ನು ಮತ್ತು ಸಂವಹನಾತ್ಮಕ ಬಳಕೆದಾರ ಅನುಭವವನ್ನು ಸುಧಾರಿಸಲು ಟೋಕನ್ ಸ್ಟ್ರೀಮಿಂಗ್ ಅನ್ನು ಪ್ರದರ್ಶಿಸುತ್ತದೆ. ಲೂಪ್ ಆಗಾಗ್ಗೆ ಬಂದಂತೆ ಹೆಚ್ಚುವರಿ ಡೆಲ್ಟಾಗಳನ್ನು ಮುದ್ರಿಸುತ್ತದೆ:
- ಆರಂಭಿಕ ಭಾಗಶಃ ಔಟ್‌ಪುಟ್ ಮುಖ್ಯವಾಗಿರುವ ಚಾಟ್ UI ಗಳಿಗೆ ಉಪಯುಕ್ತ.
- ಸಂಪೂರ್ಣ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆ ತಡತೆಯ ವಿರುದ್ಧ ಟೋಕನ್ ಥ್ರೂಪುಟ್ ಅನ್ನು ಅಳೆಯಲು ಅವಕಾಶ ನೀಡುತ್ತದೆ.

ನೀವು ಈ ಮಾದರಿಯನ್ನು ಟೋಕನ್ ಗಳನ್ನು ಸಂಗ್ರಹಿಸಲು, ಪ್ರಗತಿ ವಿದ್ಜೆಟ್ ಅನ್ನು ನವೀಕರಿಸಲು ಅಥವಾ ಮಧ್ಯದಲ್ಲಿ ರದ್ದುಗೊಳಿಸಲು ಹೊಂದಿಸಬಹುದು.


```python
# Streaming example
stream = client.chat.completions.create(
    model=model_id,
    messages=[{'role':'user','content':'Give a one-sentence definition of edge AI.'}],
    stream=True,
    max_tokens=60,
    temperature=0.4
)
for chunk in stream:
    delta = chunk.choices[0].delta
    if delta and delta.content:
        print(delta.content, end='', flush=True)
print()
```
