# бһҹбһҳбҹҗбһҷ 4 вҖ“ бһҖбһ¶бһҡбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ” SLM бһ“бһ·бһ„ LLM

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһ–бҹҒбһӣбһҷбһәбһҸ (latency) бһ“бһ·бһ„бһӮбһ»бһҺбһ—бһ¶бһ–бһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһӮбҹҶбһҡбһј бһҡбһңбһ¶бһ„бһҳбҹүбһјбһҢбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҒбҹ’бһ“бһ¶бһҸбһҸбһјбһ… бһ“бһ·бһ„бһҳбҹүбһјбһҢбҹӮбһӣбһ’бҹҶбһҳбһҪбһҷбһҠбҹӮбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҸбһ¶бһҳ Foundry Localбҹ”


## вҡЎ бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҠбһҫбһҳбһҡбһ бҹҗбһҹ

**бһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһІбҹ’бһҷбһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһ…бһ„бһ…бһ¶бҹҶ (бһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“бһ—бһ¶бһ–):**
1. бһҳбҹүбһјбһҢбҹӮбһӣбһҮбҹ’бһҡбһҫбһҹбһ”бҹ’бһҡбһ—бҹҒбһ‘ CPU бһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ· (бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһӣбһҫбһ§бһ”бһҖбһҡбһҺбҹҚбһҺбһ¶бһҳбһҪбһҷ)
2. бһ”бҹ’бһҡбһҫ `qwen2.5-3b` бһҮбҹҶбһ“бһҪбһҹ 7B (бһҹбһ“бҹ’бһҹбҹҶбһ–бһё ~4GB RAM)
3. бһҖбһ¶бһҡбһҹбҹ’бһӮбһ¶бһӣбҹӢбһ…бҹ’бһҡбһҖбһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ· (бһҳбһ·бһ“бһ‘бһ¶бһҳбһ‘бһ¶бһҡбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҠбҹ„бһҷбһҠбҹғ)
4. бһ‘бҹҶбһ бҹҶ RAM бһҹбһҡбһ»бһ”бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ: бһ”бҹ’бһҡбһ бҹӮбһӣ ~8GB (бһҺбҹӮбһ“бһ¶бҹҶ) (бһҳбҹүбһјбһҢбҹӮбһӣ + OS)

**бһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢ Terminal (30 бһңбһ·бһ“бһ¶бһ‘бһё):**
```bash
foundry service start
foundry model run phi-4-mini
foundry model run qwen2.5-3b
```

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҡбһҸбҹӢ notebook бһ“бҹҒбҹҮ! рҹҡҖ


### бһҖбһ¶бһҡбһ–бһ“бҹ’бһҷбһӣбҹӢ: бһҖбһ¶бһҡбһҸбҹҶбһЎбһҫбһ„бһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһ—бһ¶бһ–
бһҸбҹҶбһЎбһҫбһ„бһҖбһүбҹ’бһ…бһ”бҹӢбһўбһ”бҹ’бһ”бһ”бһҡбһҳбһ¶ (`foundry-local-sdk`, `openai`, `numpy`) бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһңбһ¶бһҹбҹӢбһ–бҹҒбһӣ бһ“бһ·бһ„бһҹбҹҶбһҺбһҫбһҹбһ“бҹ’бһ‘бһ“бһ¶бҹ” бһўбһ¶бһ…бһҡбһҸбҹӢбһЎбһҫбһ„бһңбһ·бһүбһ”бһ¶бһ“бһҠбҹ„бһҷбһӮбҹ’бһҳбһ¶бһ“бһ•бһӣбһ”бҹүбҹҮбһ–бһ¶бһӣбҹӢбҹ”


# бһҹбҹҒбһҺбһ¶бһҡбһёбһҷбҹ„
бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһҳбҹүбһјбһҢбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҸбһјбһ…бһҳбһҪбһҷ (SLM) бһҮбһ¶бһҳбһҪбһҷбһҳбҹүбһјбһҢбҹӮбһӣбһ’бҹҶбһҮбһ¶бһ„бһ“бҹ…бһӣбһҫбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһ…бһјбһӣбһҸбҹӮбһҳбһҪбһҷ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһё бһҖбһ¶бһҡбһҮбһҪбһүбһҠбһјбһҡвҖ‘бһўбһ¶бһ…бһҳбһ¶бһ“бҹҲ
- **бһҖбһ¶бһҡбһ”бҹ’бһҡбҹӮбһ”бҹ’бһҡбһҪбһӣбһӣбҹ’бһ”бһҝбһ“** (бһңбһ·бһ“бһ¶бһ‘бһёбһ–бҹҒбһӣбһңбҹҒбһӣбһ¶бһ–бһ·бһҸ)
- **бһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҸбһјбһҖбһ·бһ“** (бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ¶бһ“) бһҮбһ¶бһҸбҹҶбһҺбһ¶бһ„бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ‘бһ·бһ“бҹ’бһ“бһ•бһӣ
- **бһӣбһ‘бҹ’бһ’бһ•бһӣбһӮбһ»бһҺбһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҡбһ бҹҗбһҹ**
- **бһҖбһ¶бһҡбһӮбһҺбһ“бһ¶бһӣбҹ’бһ”бһҝбһ“** бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бһёбһҖбһ¶бһҡбһҖбһҫбһ“бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–

**бһўбһҗбҹҒбһҡбһ”бһҡбһ·бһ”бһ‘бһ”бһҡбһ·бһҹбҹ’бһҗбһ¶бһ“бҹ–**
- `SLM_ALIAS` - бһҳбҹүбһјбһҢбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҸбһјбһ… (бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ: phi-4-mini, бһ”бҹ’бһҡбһ бһ¶бһҖбҹӢбһ”бҹ’бһҡбһ бҹӮбһӣ 4GB RAM)
- `LLM_ALIAS` - бһҳбҹүбһјбһҢбҹӮбһӣбһ—бһ¶бһҹбһ¶бһ’бҹҶбһҮбһ¶бһ„ (бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ: qwen2.5-7b, бһ”бҹ’бһҡбһ бһ¶бһҖбҹӢбһ”бҹ’бһҡбһ бҹӮбһӣ 7GB RAM)
- `COMPARE_PROMPT` - бһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһ…бһјбһӣбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”
- `COMPARE_RETRIES` - бһ…бҹҶбһ“бһҪбһ“бһ–бҹ’бһҷбһ¶бһҷбһ¶бһҳбһЎбһҫбһ„бһңбһ·бһүбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһ–бһҡбһ№бһ„бһҳбһ¶бҹҶ (бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ: 2)
- `FOUNDRY_LOCAL_ENDPOINT` - бһ”бҹ’бһҸбһјбһҡбһ‘бһёбһҸбһ¶бҹҶбһ„бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ (бһҡбһҖбһғбһҫбһүбһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ· бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ·бһ“бһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ)

**бһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһңбһ¶бһ’бҹ’бһңбһҫбһҖбһ¶бһҡ (бһӣбҹҶбһ“бһ¶бҹҶ SDK бһ•бҹ’бһӣбһјбһңбһҖбһ¶бһҡ):**
1. **FoundryLocalManager** бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳ бһ“бһ·бһ„бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ Foundry Local
2. бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ· бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ·бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ (бһҳбһ·бһ“бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҠбҹ„бһҷбһҠбҹғ)
3. бһҳбҹүбһјбһҢбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ–бһёбһҲбҹ’бһҳбҹ„бҹҮбһҡбһ„бһ‘бҹ…бһҖбһ¶бһ“бҹӢ ID бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ·
4. бһҮбһҳбҹ’бһҡбһҫбһҹбһҖбҹҶбһҺбҹӮбһҠбҹӮбһӣбһўбһ»бһ”бһ‘бһёбһҳбҹүбҹғбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢHardwareбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҮбҹ’бһҡбһҫбһҹ (CUDA, NPU, бһ¬ CPU)
5. бһўбһҸбһ·бһҗбһ·бһҮбһ“бһҠбҹӮбһӣбһҹбҹ’бһҡбһ”бһӮбҹ’бһ“бһ¶бһҮбһ¶бһҳбһҪбһҷ OpenAI бһўбһ“бһ»бһңбһҸбҹ’бһҸбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһ”бҹӢбһҹбһ“бҹ’бһ‘аёҷаёІ
6. бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбһ¶бһҸбҹ’бһҡбһҠбҹ’бһӢбһ¶бһ“бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ…бһ¶бһ”бҹӢбһҷбһҖбҹ– бһ–бҹҒбһӣбһҷбһәбһҸ (latency), бһҸбһјбһҖбһ·бһ“, бһӮбһ»бһҺбһ—бһ¶бһ–бһӣбһ‘бҹ’бһ’бһ•бһӣ
7. бһӣбһ‘бҹ’бһ’бһ•бһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ” бһҠбһҫбһҳбҹ’бһ”бһёбһӮбһҺбһ“бһ¶бһўбһҸбҹ’бһҡбһ¶бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһ“бһӣбҹ’бһ”бһҝбһ“

бһҖбһ¶бһҡбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһҸбһјбһ…бһ“бҹҒбҹҮбһҮбһҪбһҷбһҹбһҳбҹ’бһҡбҹҒбһ…бһҗбһ¶ бһ–бҹҒбһӣбһҺбһ¶бһҠбҹӮбһӣбһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһјбһ“бһ‘бҹ…бһҳбҹүбһјбһҢбҹӮбһӣбһ’бҹҶбһҮбһ¶бһ„бһҮбһ¶бһҠбҹҶбһҺбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҳбһ¶бһ“бһ бҹҒбһҸбһ»бһ•бһӣбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”

- **Hardware Optimization** - бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҖбҹҶбһҺбҹӮбһҠбҹҸбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ§бһ”бһҖбһҡбһҺбҹҚбһҠбҹӮбһӣбһҳбһ¶бһ“ (CUDA, NPU, or CPU)
- **OpenAI Client** - бһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҮбһ¶бһҳбһҪбһҷ endpoint бһҡбһ”бһҹбҹӢ manager бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бһјбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ API бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһӮбҹ’бһ“бһ¶ OpenAI

**бһӣбһҖбҹ’бһҒбһҺбҹҲбһ’бһ“бҹӢбһ‘бҹ’бһҡбһ¶бҹҶ:**
- бһҷбһ»бһ‘бҹ’бһ’бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡвҖӢбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһЎбһҫбһ„бһңбһ·бһүбһ”бҹ’бһҡбһҫ exponential backoff (бһўбһ¶бһ…бһҖбҹҶбһҺбһҸбҹӢбһҸбһ¶бһҳбһ”бһҡбһ·бһҹбҹ’бһҗбһ¶бһ“)
- бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳбһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ· бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ·бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ
- бһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢ бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҖбһ¶бһҡбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳ
- бһҖбһ¶бһҡвҖӢбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҖбҹҶбһ бһ»бһҹвҖӢбһҷбҹүбһ¶бһ„бһ‘бһ“бҹӢбһ—бҹ’бһӣбһ“бҹӢ бһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡбһҡбһ¶бһҷбһҖбһ¶бһҡбһҺбҹҚбһҖбҹҶбһ бһ»бһҹбһӣбһҳбҹ’бһўбһ·бһҸ
- бһҖбһ¶бһҡбһҖбһҖбҹӢбһҳбҹүбһјбһҠбҹӮбһӣбһҖбҹ’бһ“бһ»бһ„ cache бһҠбһҫбһҳбҹ’бһ”бһёбһҮбҹҖбһҹбһңбһ¶бһ„бһҖбһ¶бһҡбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһЎбһҫбһ„бһңбһ·бһүбһҮбһ¶бһҳбҹ’бһҸбһ„бһ‘бҹҖбһҸ

**бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһӣбһ‘бҹ’бһ’бһ•бһӣ:**
- бһҖбһ¶бһҡбһңбһ¶бһҹбҹӢбһӣбһ¶бһҸбһ„бҹӢ (wall clock time)
- бһҸбһ¶бһҳбһҠбһ¶бһ“бһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ token (бһ”бһҫбһҳбһ¶бһ“)
- бһӣбһ‘бҹ’бһ’бһ•бһӣбһӮбҹҶбһҡбһј (бһ”бһ¶бһ“бһҖбһ¶бһҸбҹӢбһҒбҹ’бһӣбһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһ–бһ„бһ¶бһҷбһўбһ¶бһ“)
- бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһӣбһҳбҹ’бһўбһ·бһҸбһ–бһёбһҖбҹҶбһ бһ»бһҹбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбҹҶбһҺбһҫбһҠбҹӮбһӣбһ”бһҡбһ¶бһҮбҹҗбһҷ

бһӮбҹҶбһҡбһјбһ“бҹҒбҹҮбһ”бҹ’бһҡбһҫбһҳбҹүбһјбһҢбһ»бһӣ workshop_utils бһҠбҹӮбһӣбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҸбһ¶бһҳбһӮбҹҶбһҡбһј SDK бһ•бҹ’бһӣбһјбһңбһҖбһ¶бһҡбҹ”

**SDK Reference:**
- бһғбҹ’бһӣбһ¶бҹҶбһ„бһҖбһјбһҠбһ…бһҳбҹ’бһ”бһ„: https://github.com/microsoft/Foundry-Local
- SDK бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ Python: https://github.com/microsoft/Foundry-Local/tree/main/sdk/python/foundry_local
- Workshop Utils: ../samples/workshop_utils.py


````python
def setup(alias: str, endpoint: str = None, retries: int = 3):
    """
    Initialize a Foundry Local model connection using official SDK pattern.
    
    This follows the workshop_utils pattern which uses FoundryLocalManager
    to properly initialize the Foundry Local service and resolve models.
    
    Args:
        alias: Model alias (e.g., 'phi-4-mini', 'qwen2.5-3b')
        endpoint: Optional endpoint override (usually auto-detected)
        retries: Number of connection attempts (default: 3)
    
    Returns:
        tuple: (manager, client, model_id, metadata) or (None, None, alias, error_metadata) if failed
    """
    import time
    
    last_err = None
    current_delay = 2  # seconds
    
    for attempt in range(1, retries + 1):
        try:
            print(f"[Init] Connecting to '{alias}' (attempt {attempt}/{retries})...")
            
            # Use the workshop utility which follows the official SDK pattern
            manager, client, model_id = get_client(alias, endpoint=endpoint)
            
            print(f"[OK] Connected to '{alias}' -> {model_id}")
            print(f"     Endpoint: {manager.endpoint}")
            
            return manager, client, model_id, {
                'endpoint': manager.endpoint,
                'resolved': model_id,
                'attempts': attempt,
                'status': 'success'
            }
            
        except Exception as e:
            last_err = e
            error_msg = str(e)
            
            # Provide helpful error messages
            if "Connection error" in error_msg or "connection refused" in error_msg.lower():
                print(f"[ERROR] Cannot connect to Foundry Local service")
                print(f"        вҶ’ Is the service running? Try: foundry service start")
                print(f"        вҶ’ Is the model loaded? Try: foundry model run {alias}")
            elif "not found" in error_msg.lower():
                print(f"[ERROR] Model '{alias}' not found in catalog")
                print(f"        вҶ’ Available models: Run 'foundry model ls' in terminal")
                print(f"        вҶ’ Download model: Run 'foundry model download {alias}'")
            else:
                print(f"[ERROR] Setup failed: {type(e).__name__}: {error_msg}")
            
            if attempt < retries:
                print(f"[Retry] Waiting {current_delay:.1f}s before retry...")
                time.sleep(current_delay)
                current_delay *= 2  # Exponential backoff
    
    # All retries failed - provide actionable guidance
    print(f"\nвқҢ Failed to initialize '{alias}' after {retries} attempts")
    print(f"   Last error: {type(last_err).__name__}: {str(last_err)}")
    print(f"\nрҹ’Ў Troubleshooting steps:")
    print(f"   1. Ensure Foundry Local service is running:")
    print(f"      вҶ’ foundry service status")
    print(f"      вҶ’ foundry service start (if not running)")
    print(f"   2. Ensure model is loaded:")
    print(f"      вҶ’ foundry model run {alias}")
    print(f"   3. Check available models:")
    print(f"      вҶ’ foundry model ls")
    print(f"   4. Try alternative models if '{alias}' isn't available")
    
    return None, None, alias, {
        'error': f"{type(last_err).__name__}: {str(last_err)}",
        'endpoint': endpoint or 'auto-detect',
        'attempts': retries,
        'status': 'failed'
    }


def run(client, model_id: str, prompt: str, max_tokens: int = 180, temperature: float = 0.5):
    """
    Run inference with the configured model using OpenAI SDK.
    
    Args:
        client: OpenAI client instance (configured for Foundry Local)
        model_id: Model identifier (resolved from alias)
        prompt: Input prompt
        max_tokens: Maximum response tokens (default: 180)
        temperature: Sampling temperature (default: 0.5)
    
    Returns:
        dict: Response with timing, tokens, and content
    """
    import time
    
    start = time.time()
    
    try:
        response = client.chat.completions.create(
            model=model_id,
            messages=[{"role": "user", "content": prompt}],
            max_tokens=max_tokens,
            temperature=temperature
        )
        
        elapsed = time.time() - start
        
        # Extract response details
        content = response.choices[0].message.content
        
        # Try to extract token usage from multiple possible locations
        usage_info = {}
        if hasattr(response, 'usage') and response.usage:
            usage_info['prompt_tokens'] = getattr(response.usage, 'prompt_tokens', None)
            usage_info['completion_tokens'] = getattr(response.usage, 'completion_tokens', None)
            usage_info['total_tokens'] = getattr(response.usage, 'total_tokens', None)
        
        # Calculate approximate token count if API doesn't provide it
        # Rough estimate: ~4 characters per token for English text
        if not usage_info.get('total_tokens'):
            estimated_prompt_tokens = len(prompt) // 4
            estimated_completion_tokens = len(content) // 4
            estimated_total = estimated_prompt_tokens + estimated_completion_tokens
            usage_info['estimated_tokens'] = estimated_total
            usage_info['estimated_prompt_tokens'] = estimated_prompt_tokens
            usage_info['estimated_completion_tokens'] = estimated_completion_tokens
        
        return {
            'status': 'success',
            'content': content,
            'elapsed_sec': elapsed,
            'tokens': usage_info.get('total_tokens') or usage_info.get('estimated_tokens'),
            'usage': usage_info,
            'model': model_id
        }
        
    except Exception as e:
        elapsed = time.time() - start
        return {
            'status': 'error',
            'error': f"{type(e).__name__}: {str(e)}",
            'elapsed_sec': elapsed,
            'model': model_id
        }


print("вң… Execution helpers defined: setup(), run()")
print("   вҶ’ Uses workshop_utils for proper SDK integration")
print("   вҶ’ setup() initializes with FoundryLocalManager")
print("   вҶ’ run() executes inference via OpenAI-compatible API")
print("   вҶ’ Token counting: Uses API data or estimates if unavailable")
````

бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҠбһҫбһҳбһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹҶбһ бҹҶбһҹбҹ’бһҡбһ¶бһӣбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ FoundryLocalManager бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбҹүбһјбһҢбҹӮбһӣбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбҹ” бһ“бҹҒбҹҮбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҗбһ¶бҹ–

- бһҲбҹ’бһҳбҹ„бҹҮбһҮбҹҶбһ“бһҪбһҹбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ‘бҹ…бһҮбһ¶ ID бһҳбҹүбһјбһҢбҹӮбһӣбһ–бһ·бһҸбһ”бҹ’бһҡбһ¶бһҖбһҠ
- бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҳбһ¶бһ“бһҹбҹ’бһҗбһ·бһҡбһ—бһ¶бһ–бһҳбһ»бһ“бһ–бҹҒбһӣбһҡбһҸбҹӢбһҖбһ¶бһҡбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”

бһҳбһ»бһҒбһ„бһ¶бһҡ setup() бһ”бҹ’бһҡбһҫбһӣбҹҶбһ“бһ¶бҹҶ SDK бһ•бҹ’бһӣбһјбһңбһҖбһ¶бһҡбһ–бһё workshop_utilsбҹ”


````python
# Simplified diagnostic: Just verify service is accessible
import requests

def check_foundry_service():
    """Quick diagnostic to verify Foundry Local is running."""
    # Try common ports
    endpoints_to_try = [
        "http://localhost:59959",
        "http://127.0.0.1:59959", 
        "http://localhost:55769",
        "http://127.0.0.1:55769",
    ]
    
    print("[Diagnostic] Checking Foundry Local service...")
    
    for endpoint in endpoints_to_try:
        try:
            response = requests.get(f"{endpoint}/health", timeout=2)
            if response.status_code == 200:
                print(f"вң… Service is running at {endpoint}")
                
                # Try to list models
                try:
                    models_response = requests.get(f"{endpoint}/v1/models", timeout=2)
                    if models_response.status_code == 200:
                        models_data = models_response.json()
                        model_count = len(models_data.get('data', []))
                        print(f"вң… Found {model_count} models available")
                        if model_count > 0:
                            print("   Models:", [m.get('id', 'unknown') for m in models_data.get('data', [])[:5]])
                except Exception as e:
                    print(f"вҡ пёҸ  Could not list models: {e}")
                
                return endpoint
        except requests.exceptions.ConnectionError:
            continue
        except Exception as e:
            print(f"вҡ пёҸ  Error checking {endpoint}: {e}")
    
    print("\nвқҢ Foundry Local service not found!")
    print("\nрҹ’Ў To fix this:")
    print("   1. Open a terminal")
    print("   2. Run: foundry service start")
    print("   3. Run: foundry model run phi-4-mini")
    print("   4. Run: foundry model run qwen2.5-3b")
    print("   5. Re-run this notebook")
    return None

# Run diagnostic
discovered_endpoint = check_foundry_service()

if discovered_endpoint:
    print(f"\nвң… Service detected (will be managed by FoundryLocalManager)")
else:
    print(f"\nвҡ пёҸ  No service detected - FoundryLocalManager will attempt to start it")
````

бһ”бһҫбһўбҹ’бһ“бһҖбһғбһҫбһү `APIConnectionError`, бһҹбҹҒбһңбһ¶ Foundry Local бһўбһ¶бһ…бһҖбҹҶбһ–бһ»бһ„бһҳбһ·бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ бһ¬бһҳбҹүбһјбһҢбҹӮбһӣбһҳбһ·бһ“бһ‘бһ¶бһ“бҹӢбһ”бһ¶бһ“бһ•бҹ’бһ‘бһ»бһҖбҹ” бһҹбһјбһҳбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҮбҹҶбһ бһ¶бһ“бһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбҹ–

**1. бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҹбҹҒбһңбһ¶:**
```bash
# бһ“бҹ…вҖӢбһҖбҹ’бһ“бһ»бһ„вҖӢбһ•бҹ’бһ‘бһ¶бҹҶбһ„вҖӢбһ”бһүбҹ’бһҮбһ¶ (бһҳбһ·бһ“вҖӢбһҳбҹӮбһ“вҖӢбһ“бҹ…вҖӢбһҖбҹ’бһ“бһ»бһ„вҖӢбһҹбҹҖбһңбһ—бҹ…вҖӢбһҖбҹҶбһҺбһҸбҹӢвҖӢбһҸбҹ’бһҡбһ¶):
foundry service status
```

**2. бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҹбҹҒбһңбһ¶ (бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ·бһ“бһҖбҹҶбһ–бһ»бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ):**
```bash
foundry service start
```

**3. бһ•бҹ’бһ‘бһ»бһҖбһҳбҹүбһјбһҢбҹӮбһӣбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ:**
```bash
# бһ•бҹ’бһ‘бһ»бһҖбһҳбҹүбһјбһҢбҹӮбһӣбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”
foundry model run phi-4-mini
foundry model run qwen2.5-7b

# бһ¬бһ”бҹ’бһҡбһҫбһҳбҹүбһјбһҢбҹӮбһӣбһҮбҹҶбһ“бһҪбһҹ:
foundry model run phi-3.5-mini
foundry model run qwen2.5-3b
```

**4. бһ”бһүбҹ’бһ…бһ¶бһҖбҹӢбһҗбһ¶бһҳбҹүбһјбһҢбҹӮбһӣбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“:**
```bash
foundry model ls
```

**бһ”бһүбҹ’бһ бһ¶бһ‘бһјбһ‘бҹ…:**
- вқҢ бһҹбҹҒбһңбһ¶бһҳбһ·бһ“бһҖбҹҶбһ–бһ»бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ вҶ’ бһҡбһҸбҹӢ `foundry service start`
- вқҢ бһҳбҹүбһјбһҢбҹӮбһӣбһҳбһ·бһ“бһ‘бһ¶бһ“бҹӢбһ”бһ¶бһ“бһ•бҹ’бһ‘бһ»бһҖ вҶ’ бһҡбһҸбҹӢ `foundry model run <model-name>`
- вқҢ бһҮбһҳбҹ’бһӣбҹ„бҹҮбһ…бҹ’бһҡбһҖ вҶ’ бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһҹбҹҒбһңбһ¶бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһҖбҹҶбһ–бһ»бһ„бһ”бҹ’бһҡбһҫбһ…бҹ’бһҡбһҖбһ“бҹ„бҹҮбһ¬бһ‘бҹҒ
- вқҢ бһҖбһ¶бһҡбһҡбһ¶бҹҶбһ„бһҒбҹ’бһ‘бһ”бҹӢбһҠбҹ„бһҷ Firewall вҶ’ бһ’бһ¶бһ“бһ¶бһҗбһ¶бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҖбҹ’бһ“бһ»бһ„бһҹбҹ’бһҡбһ»бһҖбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸ

**бһҠбҹҶбһҺбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһӣбһҝбһ“:** бһҡбһҸбҹӢбһҖбҹ„бһҹбһ·бһҖбһ¶бһ–бһ·бһ“бһ·бһҸбҹ’бһҷ (diagnostic cell) бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ бһҳбһ»бһ“бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһ»бһ“


````python
preflight = {}
retries = 2  # Number of retry attempts

for a in (SLM, LLM):
    mgr, c, mid, info = setup(a, endpoint=ENDPOINT, retries=retries)
    # Keep the original status from info (either 'success' or 'failed')
    preflight[a] = info

print('\n[Pre-flight Check]')
for alias, details in preflight.items():
    status_icon = 'вң…' if details['status'] == 'success' else 'вқҢ'
    print(f"  {status_icon} {alias}: {details['status']} - {details.get('resolved', details.get('error', 'unknown'))}")

preflight
````

### вң… бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһ»бһ“бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҠбһҫбһҳбҹ– бһ—бһ¶бһ–бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“бһ“бҹғбһҳбҹүбһјбһҠбҹӮбһӣ

бһҖбҹ„бһҹбһ·бһҖбһ¶бһ“бҹҒбҹҮбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһҗбһ¶ бһҳбҹүбһјбһҠбҹӮбһӣбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһўбһ¶бһ…бһ…бһјбһӣбһҠбһӣбҹӢбһ“бҹ…бһ…бҹҶбһҺбһ»бһ…бһ”бһүбҹ’бһ…бһ”бҹӢбһҠбҹӮбһӣбһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ бһҳбһ»бһ“бһ–бҹҒбһӣбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бҹ”


### бһҖбһ¶бһҡбһ–бһ“бҹ’бһҷбһӣбҹӢ: бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһҖбһ¶бһҡбһҡбһҸбҹӢ & бһ”бҹ’бһҡбһҳбһјбһӣбһӣбһ‘бҹ’бһ’бһ•бһӣ
бһ’бҹ’бһңбһҫбһңбһ·бһӣбһӣбһҫ aliases бһ‘бһ¶бҹҶбһ„бһ–бһёбһҡ бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһӣбҹҶбһ“бһ¶бҹҶ Foundry SDK бһ•бҹ’бһӣбһјбһңбһҖбһ¶бһҡ:
1. бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҳбҹүбһјбһҢбҹӮбһӣбһ“бһёбһҳбһҪбһҷбҹ—бһҮбһ¶бһҳбһҪбһҷ setup() (бһ”бҹ’бһҡбһҫ FoundryLocalManager)
2. бһҡбһҸбҹӢ inference бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ OpenAI-compatible API
3. бһҖбһҸбҹӢбһҸбҹ’бһҡбһ¶ latency, tokens, бһ“бһ·бһ„ sample output
4. бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҹбһ„бҹ’бһҒбҹҒбһ” JSON бһҮбһ¶бһҳбһҪбһҷбһңбһ·бһ—бһ¶бһӮбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”

бһ“бҹҒбҹҮбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҸбһ¶бһҳбһӣбҹҶбһ“бһ¶бҹҶбһҠбһјбһ…бһӮбҹ’бһ“бһ¶бһ“бҹғбһӮбҹҶбһҡбһј Workshop бһҖбҹ’бһ“бһ»бһ„ session04/model_compare.py.


````python
results = []
retries = 2  # Number of retry attempts

for alias in (SLM, LLM):
    mgr, client, mid, info = setup(alias, endpoint=ENDPOINT, retries=retries)
    if client:
        r = run(client, mid, PROMPT)
        results.append({'alias': alias, **r})
    else:
        # If setup failed, record error
        results.append({
            'alias': alias,
            'status': 'error',
            'error': info.get('error', 'Setup failed'),
            'elapsed_sec': 0,
            'tokens': None,
            'model': alias
        })

# Display results
print(json.dumps(results, indent=2))

# Quick comparative view
print('\n' + '='*80)
print('COMPARISON SUMMARY')
print('='*80)
print(f"{'Alias':<20} {'Status':<15} {'Latency(s)':<15} {'Tokens':<15}")
print('-'*80)

for row in results:
    status = row.get('status', 'unknown')
    status_icon = 'вң…' if status == 'success' else 'вқҢ'
    latency_str = f"{row.get('elapsed_sec', 0):.3f}" if row.get('elapsed_sec') else 'N/A'
    
    # Handle token display - show if available or indicate estimated
    tokens = row.get('tokens')
    usage = row.get('usage', {})
    if tokens:
        if 'estimated_tokens' in usage:
            tokens_str = f"~{tokens} (est.)"
        else:
            tokens_str = str(tokens)
    else:
        tokens_str = 'N/A'
    
    print(f"{status_icon} {row['alias']:<18} {status:<15} {latency_str:<15} {tokens_str:<15}")

print('-'*80)

# Show detailed token breakdown if available
print("\nDetailed Token Usage:")
for row in results:
    if row.get('status') == 'success' and row.get('usage'):
        usage = row['usage']
        print(f"\n  {row['alias']}:")
        if 'prompt_tokens' in usage and usage['prompt_tokens']:
            print(f"    Prompt tokens:     {usage['prompt_tokens']}")
            print(f"    Completion tokens: {usage['completion_tokens']}")
            print(f"    Total tokens:      {usage['total_tokens']}")
        elif 'estimated_tokens' in usage:
            print(f"    Estimated prompt:     {usage['estimated_prompt_tokens']}")
            print(f"    Estimated completion: {usage['estimated_completion_tokens']}")
            print(f"    Estimated total:      {usage['estimated_tokens']}")
            print(f"    (API did not provide token counts - using ~4 chars/token estimate)")

print('\n' + '='*80)

# Calculate speedup if both succeeded
if len(results) == 2 and all(r.get('status') == 'success' and r.get('elapsed_sec') for r in results):
    speedup = results[1]['elapsed_sec'] / results[0]['elapsed_sec']
    print(f"\nрҹ’Ў SLM is {speedup:.2f}x faster than LLM for this prompt")
    
    # Compare token throughput if available
    slm_tokens = results[0].get('tokens', 0)
    llm_tokens = results[1].get('tokens', 0)
    if slm_tokens and llm_tokens:
        slm_tps = slm_tokens / results[0]['elapsed_sec']
        llm_tps = llm_tokens / results[1]['elapsed_sec']
        print(f"   SLM throughput: {slm_tps:.1f} tokens/sec")
        print(f"   LLM throughput: {llm_tps:.1f} tokens/sec")
        
elif any(r.get('status') == 'error' for r in results):
    print(f"\nвҡ пёҸ  Some models failed - check errors above")
    print("   Ensure Foundry Local is running: foundry service start")
    print("   Ensure models are loaded: foundry model run <model-name>")

results
````

### бһҖбһ¶бһҡбһ”бһҖбһҹбҹ’бһҡбһ¶бһҷбһӣбһ‘бҹ’бһ’бһ•бһӣ

**бһҳбһ¶бһҸбҹ’бһҡбһҠбҹ’бһӢбһ¶бһ“бһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ—:**
- **Latency**: бһҸбһ·бһ…бһӣбҹ’бһўбһҮбһ¶бһ„ - бһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһ–бҹҒбһӣбһҶбҹ’бһӣбһҫбһҷбһӣбһҝбһ“бһҮбһ¶бһ„
- **Tokens**: бһўбһҸбҹ’бһҡбһ¶бһ”бһүбҹ’бһ…бҹҒбһүбһҒбҹ’бһ–бһҹбҹӢ = бһҳбһ¶бһ“ token бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡвҖӢбһ…бҹ’бһҡбһҫбһ“
- **Route**: бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҗбһ¶ API endpoint бһҺбһ¶ бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫ

**бһ–бҹҒбһӣбһҺбһ¶бһӮбһҪбһҡбһ”бҹ’бһҡбһҫ SLM бһ“бһ·бһ„ LLM:**
- **SLM (бһҳбҹүбһјбһҢбҹӮбһӣбһ—бһ¶бһҹбһ¶бһ‘бҹҶбһ бҹҶбһҸбһјбһ…)**: бһҶбҹ’бһӣбһҫбһҷбһҶбһ¶бһ”бҹӢ, бһ”бҹ’бһҡбһҫбһ’бһ“бһ’бһ¶бһ“бһҸбһ·бһ…, бһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҡбһҖбһ·бһ…бҹ’бһ…бһҹбһ¶бһҳбһүбҹ’бһү
- **LLM (бһҳбҹүбһјбһҢбҹӮбһӣбһ—бһ¶бһҹбһ¶бһ‘бҹҶбһ бҹҶбһ’бҹҶ)**: бһӮбһ»бһҺбһ—бһ¶бһ–бһҒбҹ’бһ–бһҹбҹӢ, бһҖбһ¶бһҡбһӮбһ·бһҸбһңбһ·бһ—бһ¶бһӮбһ”бҹ’бһҡбһҹбһҫбһҡ, бһ”бҹ’бһҡбһҫбһ“бҹ…бһ–бҹҒбһӣбһҠбҹӮбһӣбһӮбһ»бһҺбһ—бһ¶бһ–бһҳбһ¶бһ“бһҹбһ¶бһҡбҹҲбһҹбҹҶбһҒбһ¶бһ“бҹӢ

**бһҮбҹҶбһ бһ¶бһ“бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢ:**
1. бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„ prompts бһ•бҹ’бһҹбҹҒбһ„бҹ— бһҠбһҫбһҳбҹ’бһ”бһёбһҳбһҫбһӣбһҗбһ¶ бһ—бһ¶бһ–бһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһү бһ”бҹүбҹҮбһ–бһ¶бһӣбҹӢбһҠбһӣбҹӢбһҖбһ¶бһҡбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…
