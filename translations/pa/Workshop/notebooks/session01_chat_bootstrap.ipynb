# ਸੈਸ਼ਨ 1 – ਚੈਟ ਬੂਟਸਟਰੈਪ (Foundry Local)

ਇਹ ਨੋਟਬੁੱਕ Foundry Local ਨੂੰ ਬੂਟਸਟਰੈਪ ਕਰਦੀ ਹੈ, ਪਸੰਦੀਦਾ ਮਾਡਲ ਅਲਿਆਸ ਡਾਊਨਲੋਡ ਕਰਦੀ ਹੈ, ਅਤੇ ਮਿਆਰੀ ਅਤੇ ਸਟ੍ਰੀਮਿੰਗ ਚੈਟ ਪੂਰਨਤਾ ਦੋਵੇਂ ਕਰਦੀ ਹੈ।


# ਸਥਿਤੀ
ਇਸ ਸੈਸ਼ਨ ਵਿੱਚ ਇੱਕ ਛੋਟੇ ਭਾਸ਼ਾ ਮਾਡਲ ਨੂੰ Foundry Local ਰਾਹੀਂ ਜਵਾਬ ਦੇਣ ਲਈ ਬਿਲਕੁਲ ਘੱਟੋ-ਘੱਟ ਜਾਣਕਾਰੀ ਦਿੱਤੀ ਜਾਂਦੀ ਹੈ। ਤੁਸੀਂ ਇਹ ਕਰੋਗੇ:
- SDK / ਕਲਾਇੰਟ dependencies ਨੂੰ ਇੰਸਟਾਲ ਕਰੋ।
- ਚੁਣੇ ਗਏ alias (ਡਿਫਾਲਟ: `phi-4-mini`) ਲਈ Foundry Local ਮੈਨੇਜਰ ਨੂੰ ਸ਼ੁਰੂ ਕਰੋ।
- ਮਾਡਲ metadata ਵਿੱਚ ਵਿਕਲਪਿਕ ਫੀਲਡਸ ਨੂੰ ਸਹਿਣ ਕਰਨ ਲਈ ਇੱਕ ਰੱਖਿਆਤਮਕ ਮੰਕੀ‑ਪੈਚ ਲਾਗੂ ਕਰੋ।
- ਇੱਕ ਸਟੈਂਡਰਡ ਚੈਟ completion ਰਿਕਵੈਸਟ ਭੇਜੋ।
- ਟੋਕਨ‑ਦਰ‑ਟੋਕਨ ਜਵਾਬ ਨੂੰ ਸਟ੍ਰੀਮ ਕਰੋ।

ਉਦੇਸ਼ ਇਹ ਹੈ ਕਿ RAG, ਰੂਟਿੰਗ ਜਾਂ ਏਜੰਟਸ 'ਤੇ ਜਾਣ ਤੋਂ ਪਹਿਲਾਂ ਤੁਹਾਡੇ ਸਥਾਨਕ runtime ਅਤੇ ਨੈਟਵਰਕ ਪਾਥ ਨੂੰ ਵੈਰੀਫਾਈ ਕੀਤਾ ਜਾਵੇ।


### ਵਿਆਖਿਆ: ਡਿਪੈਂਡੈਂਸੀ ਇੰਸਟਾਲੇਸ਼ਨ
ਇਹ ਘੱਟੋ-ਘੱਟ ਚੈਟ ਫਲੋ ਲਈ ਲੋੜੀਂਦੇ Python ਪੈਕੇਜ ਇੰਸਟਾਲ ਕਰਦਾ ਹੈ:
- `foundry-local-sdk`: ਸਥਾਨਕ ਮਾਡਲ ਅਤੇ ਸੇਵਾ ਲਾਈਫਸਾਈਕਲ ਨੂੰ ਪ੍ਰਬੰਧਿਤ ਕਰਨ ਲਈ।
- `openai`: ਚੈਟ ਪੂਰਨਤਾ ਲਈ ਜਾਣ-ਪਛਾਣ ਵਾਲਾ ਕਲਾਇੰਟ ਐਬਸਟਰੈਕਸ਼ਨ।
- `rich`: ਨੋਟਬੁੱਕ ਆਉਟਪੁੱਟ ਨੂੰ ਵਧੀਆ ਤਰੀਕੇ ਨਾਲ ਦਿਖਾਉਣ ਲਈ।

ਇਸਨੂੰ ਮੁੜ ਚਲਾਉਣਾ ਸੁਰੱਖਿਅਤ ਹੈ (idempotent)। ਜੇ ਤੁਹਾਡੇ ਵਾਤਾਵਰਣ ਵਿੱਚ ਇਹ ਪਹਿਲਾਂ ਹੀ ਮੌਜੂਦ ਹਨ ਤਾਂ ਇਸਨੂੰ ਛੱਡੋ।

### ਵਿਆਖਿਆ: ਕੋਰ ਇੰਪੋਰਟਸ  
ਨੋਟਬੁੱਕ ਵਿੱਚ ਵਰਤੇ ਜਾਣ ਵਾਲੇ ਮੋਡੀਊਲ ਲਿਆਉਂਦਾ ਹੈ:  
- `FoundryLocalManager` ਸਥਾਨਕ ਮਾਡਲ ਰਨਟਾਈਮ ਨਾਲ ਇੰਟਰਫੇਸ ਕਰਨ ਲਈ।  
- `OpenAI` ਕਲਾਇੰਟ ਤਾਂ ਜੋ ਅਸੀਂ ਜਾਣ-ਪਛਾਣ ਵਾਲੇ ਚੈਟ ਕੰਪਲੀਸ਼ਨ API ਸਤਹ ਨੂੰ ਦੁਬਾਰਾ ਵਰਤ ਸਕੀਏ।  
- `rich.print` ਸਟਾਈਲਡ ਆਉਟਪੁੱਟ ਲਈ।  

ਇੱਥੇ ਕੋਈ ਨੈਟਵਰਕ ਕਾਲ ਨਹੀਂ ਹੁੰਦੀ—ਇਹ ਸਿਰਫ ਨੇਮਸਪੇਸ ਤਿਆਰ ਕਰਦਾ ਹੈ।  


```python
import os
from foundry_local import FoundryLocalManager
from foundry_local.models import FoundryModelInfo
from openai import OpenAI
from rich import print
```

### ਵਿਆਖਿਆ: ਮੈਨੇਜਰ ਸ਼ੁਰੂਆਤ ਅਤੇ ਮੈਟਾਡੇਟਾ ਪੈਚ
ਚੁਣੇ ਗਏ ਉਪਨਾਮ ਲਈ `FoundryLocalManager` ਨੂੰ ਸ਼ੁਰੂ ਕਰਦਾ ਹੈ ਅਤੇ ਇੱਕ ਰੱਖਿਆਤਮਕ ਮੰਕੀ-ਪੈਚ ਲਾਗੂ ਕਰਦਾ ਹੈ ਤਾਂ ਜੋ ਸੇਵਾ ਦੇ ਜਵਾਬਾਂ ਨੂੰ ਸੁਚਾਰੂ ਢੰਗ ਨਾਲ ਸੰਭਾਲਿਆ ਜਾ ਸਕੇ ਜਿੱਥੇ `promptTemplate` ਸ਼ਾਇਦ `null` ਹੋ ਸਕਦਾ ਹੈ।

ਮੁੱਖ ਨਤੀਜੇ:
- ਸੇਵਾ ਦੀ ਸਥਿਤੀ ਅਤੇ ਐਂਡਪੌਇੰਟ ਦੀ ਪੁਸ਼ਟੀ ਕਰਦਾ ਹੈ।
- ਕੈਸ਼ ਕੀਤੇ ਮਾਡਲਾਂ ਦੀ ਸੂਚੀ ਦਿਖਾਉਂਦਾ ਹੈ (ਸਥਾਨਕ ਸਟੋਰ ਦੀ ਪੁਸ਼ਟੀ ਕਰਦਾ ਹੈ)।
- ਉਪਨਾਮ ਲਈ ਠੋਸ ਮਾਡਲ ID ਨੂੰ ਹੱਲ ਕਰਦਾ ਹੈ (ਜੋ ਬਾਅਦ ਦੇ ਚੈਟ ਕਾਲਾਂ ਵਿੱਚ ਵਰਤੀ ਜਾਂਦੀ ਹੈ)।

ਜੇਕਰ ਤੁਹਾਨੂੰ ਕੱਚੇ ਸੇਵਾ ਮੈਟਾਡੇਟਾ ਵਿੱਚ ਵੈਧਤਾ ਸਮੱਸਿਆਵਾਂ ਦਾ ਸਾਹਮਣਾ ਕਰਨਾ ਪਵੇ, ਤਾਂ ਇਹ ਪੈਟਰਨ ਦਿਖਾਉਂਦਾ ਹੈ ਕਿ SDK ਨੂੰ ਫੋਰਕ ਕੀਤੇ ਬਿਨਾਂ ਕਿਵੇਂ ਸਾਫ਼ ਕੀਤਾ ਜਾ ਸਕਦਾ ਹੈ।


```python
# Monkeypatch to tolerate service responses where promptTemplate is null
_original_from_list_response = FoundryModelInfo.from_list_response

def _safe_from_list_response(response):  # type: ignore
    try:
        if isinstance(response, dict) and response.get("promptTemplate") is None:
            # Normalize to empty dict so pydantic validation passes
            response["promptTemplate"] = {}
    except Exception as e:  # pragma: no cover
        print(f"[yellow]Warning: safe wrapper encountered issue normalizing promptTemplate: {e}[/yellow]")
    return _original_from_list_response(response)

# Apply patch only once
if getattr(FoundryModelInfo.from_list_response, "__name__", "") != "_safe_from_list_response":
    FoundryModelInfo.from_list_response = staticmethod(_safe_from_list_response)  # type: ignore

ALIAS = os.getenv('FOUNDRY_LOCAL_ALIAS', 'phi-4-mini')
manager = FoundryLocalManager(ALIAS)
print(f'[bold green]Service running:[/bold green] {manager.is_service_running()}')
print(f'Endpoint: {manager.endpoint}')
print('Cached models:', manager.list_cached_models())
model_id = manager.get_model_info(ALIAS).id
print(f'Using model id: {model_id}')
```

### ਵਿਆਖਿਆ: ਬੇਸਿਕ ਚੈਟ ਕੰਪਲੀਸ਼ਨ
ਇੱਕ `OpenAI`-ਅਨੁਕੂਲ ਕਲਾਇੰਟ ਬਣਾਉਂਦਾ ਹੈ ਜੋ ਸਥਾਨਕ Foundry ਐਂਡਪੌਇੰਟ ਵੱਲ ਇਸ਼ਾਰਾ ਕਰਦਾ ਹੈ ਅਤੇ ਇੱਕ ਗੈਰ-ਸਟ੍ਰੀਮਿੰਗ ਚੈਟ ਕੰਪਲੀਸ਼ਨ ਕਰਦਾ ਹੈ। ਇੱਥੇ ਧਿਆਨ ਦਿਓ:
- ਯਕੀਨੀ ਬਣਾਓ ਕਿ ਮਾਡਲ ਬਿਨਾ ਗਲਤੀ ਦੇ ਜਵਾਬ ਦਿੰਦਾ ਹੈ।
- ਲੈਟੈਂਸੀ / ਆਉਟਪੁੱਟ ਫਾਰਮੈਟ ਦੀ ਪੁਸ਼ਟੀ ਕਰੋ।
- ਸਰੋਤਾਂ ਦੀ ਬਚਤ ਕਰਨ ਲਈ `max_tokens` ਨੂੰ ਨਿਮਰ ਰੱਖੋ।

ਜੇ ਇਹ ਫੇਲ੍ਹ ਹੁੰਦਾ ਹੈ, ਤਾਂ ਦੁਬਾਰਾ ਜਾਂਚ ਕਰੋ ਕਿ Foundry Local ਸੇਵਾ ਚੱਲ ਰਹੀ ਹੈ ਅਤੇ ਐਲਿਆਸ ਸਹੀ ਤਰੀਕੇ ਨਾਲ ਹੱਲ ਹੁੰਦਾ ਹੈ।


```python
client = OpenAI(base_url=manager.endpoint, api_key=manager.api_key or 'not-needed')
prompt = 'List two benefits of local inference for privacy.'
resp = client.chat.completions.create(
    model=model_id,
    messages=[{'role':'user','content':prompt}],
    max_tokens=120,
    temperature=0.5
)
print(resp.choices[0].message.content)
```

### ਵਿਆਖਿਆ: ਸਟ੍ਰੀਮਿੰਗ ਚੈਟ ਕੰਪਲੀਸ਼ਨ  
ਟੋਕਨ ਸਟ੍ਰੀਮਿੰਗ ਨੂੰ ਬਿਹਤਰ ਮਹਿਸੂਸ ਕੀਤੀ ਜਾਣ ਵਾਲੀ ਲੈਟੈਂਸੀ ਅਤੇ ਇੰਟਰਐਕਟਿਵ UX ਲਈ ਦਰਸਾਉਂਦਾ ਹੈ। ਲੂਪ ਆਉਣ ਵਾਲੇ ਇਨਕ੍ਰਿਮੈਂਟਲ ਡੈਲਟਾ ਨੂੰ ਪ੍ਰਿੰਟ ਕਰਦਾ ਹੈ:  
- ਚੈਟ UI ਲਈ ਲਾਭਦਾਇਕ ਜਿੱਥੇ ਸ਼ੁਰੂਆਤੀ ਅਧੂਰਾ ਆਉਟਪੁੱਟ ਮਹੱਤਵਪੂਰਨ ਹੁੰਦਾ ਹੈ।  
- ਤੁਹਾਨੂੰ ਟੋਕਨ ਥਰੂਪੁੱਟ ਨੂੰ ਪੂਰੀ ਕੰਪਲੀਸ਼ਨ ਲੈਟੈਂਸੀ ਦੇ ਮੁਕਾਬਲੇ ਮਾਪਣ ਦੀ ਆਗਿਆ ਦਿੰਦਾ ਹੈ।  

ਤੁਸੀਂ ਇਸ ਪੈਟਰਨ ਨੂੰ ਟੋਕਨ ਇਕੱਠੇ ਕਰਨ, ਪ੍ਰਗਤੀ ਵਿਜਟ ਨੂੰ ਅਪਡੇਟ ਕਰਨ ਜਾਂ ਮਿਡ-ਜਨਰੇਸ਼ਨ ਨੂੰ ਰੱਦ ਕਰਨ ਲਈ ਅਨੁਕੂਲ ਕਰ ਸਕਦੇ ਹੋ।  


```python
# Streaming example
stream = client.chat.completions.create(
    model=model_id,
    messages=[{'role':'user','content':'Give a one-sentence definition of edge AI.'}],
    stream=True,
    max_tokens=60,
    temperature=0.4
)
for chunk in stream:
    delta = chunk.choices[0].delta
    if delta and delta.content:
        print(delta.content, end='', flush=True)
print()
```
